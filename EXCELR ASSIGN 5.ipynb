# Handling missing values

```python
#  Impute missing values with the median of each column
df.fillna(df.median(), inplace=True)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df['ALTV'])
```

```python
sns.boxplot(df['LB'])
```

# Outlier detection and treatment

```python

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
sns.boxplot(df['LB'])
```

```python
sns.boxplot(df['ALTV'])
```

# 4. Pattern Recognition and Insights

```python
# Correlation matrix
corr_matrix=df.corr()
corr_matrix
```

```python
# Identify and discuss strong correlations
strong_corr = corr_matrix[(corr_matrix > 0.7) | (corr_matrix < -0.7)]
print("Strong correlations: \n", strong_corr)
```
